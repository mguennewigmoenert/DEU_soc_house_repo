```python
# Finding adjacent neighbors 
gdf["neighbors"] = None
```

```python
for index, PLR_ID in gdf.iterrows():
    # get 'not disjoint' countries
    neighbors = gdf[~gdf.geometry.disjoint(PLR_ID.geometry)].PLR_NAME2.tolist()
     # remove own name of the country from the list
    neighbors = [name for name in neighbors if PLR_ID.PLR_NAME2!= name ]
    # add names of neighbors as NEIGHBORS value
    gdf.at[index, "neighbors"] = ", ".join(neighbors)

gdf = gdf.reset_index(drop=True)
gdf['PLR_ID'] = gdf['PLR_NAME2']
gdf = gdf[['PLR_ID','neighbors']]
gdf.head()
```
